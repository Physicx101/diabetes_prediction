```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn import tree
```

```python
df = pd.read_excel('diabetes_india.xlsx')
df.head()
```

```python
# PEMBUATAN VARIABEL X

X = df.drop('Outcome', axis=1)
X.head()
```

```python
# PEMBUATAN VARIABEL Y

y = df[['Outcome']]
y.head()
```

```python
# BUAT TEST DAN TRAIN SET MENGGUNAKAN STRATIFY METHOD

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, stratify=y)
X_train.head()
```

```python
# defining parameter range
parameters_svm = {'C': [0.1,1, 10], 'gamma': [1,0.1,0.01],'kernel': ['rbf','sigmoid']}
 
grid_svm = GridSearchCV(SVC(), parameters_svm, refit = True, verbose = 3)
 
# fitting the model for grid search
grid_svm.fit(X_train, y_train.values.ravel())
```

```python
parameters_gbc = {
    "n_estimators":[5,50,250],
    "max_depth":[1,3,5],
    "learning_rate":[0.1,1,10]
}

grid_gbc = GridSearchCV(GradientBoostingClassifier(), parameters_gbc, refit = True, verbose = 3)
 
# fitting the model for grid search
grid_gbc.fit(X_train, y_train.values.ravel())
```

```python
parameters_tree ={"splitter":["best","random"],
            "max_depth" : [1,3,5],
           "min_samples_leaf":[2,3,48],
           "min_weight_fraction_leaf":[0.2,0.5]}

grid_tree = GridSearchCV(tree.DecisionTreeClassifier(), parameters_tree, refit = True, verbose = 3)
 
# fitting the model for grid search
grid_tree.fit(X_train, y_train.values.ravel())
```

```python
print(grid_svm.best_params_)
print(grid_svm.best_estimator_)
grid_svm_predictions = grid_svm.predict(X_test)
```

```python
print(grid_gbc.best_params_)
print(grid_gbc.best_estimator_)
grid_gbc_predictions = grid_gbc.predict(X_test)
```

```python
print(grid_tree.best_params_)
print(grid_tree.best_estimator_)
grid_tree_predictions = grid_tree.predict(X_test)
```

```python
print(classification_report(y_test, grid_svm_predictions))
```

```python
print(classification_report(y_test, grid_gbc_predictions))
```

```python
print(classification_report(y_test, grid_tree_predictions))
```
